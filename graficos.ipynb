```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import gamma
import pandas as pd

nome = "BubbleSort"
tipo = "comp"
df = pd.read_csv(f"./{nome}_metrics/{nome}_{tipo}_metrics.csv")

# Generate x values
x = np.linspace(1, 30000, 150)  # 1000 points between 1 and 20 for smoothness

# Calculate the factorial of each x value
y = [((i**2)/1.5) for i in x]
y2 = [((i**2)/1.8) for i in x]
y_ord = df["Ordered"]
y_rev = df["Reversed"]
y_ale = df["Randomized"]

# Plotting
plt.plot(x, y, marker='', linestyle='-', color='b', label='Terica')  # Blue line
plt.plot(x, y_ord, marker='', linestyle='-', color='r', label='Ordenado')  # Red line
plt.plot(x, y2, marker='', linestyle='-', color='purple', label='Reverso')  # Red line
plt.plot(x, y_ale, marker='', linestyle='-', color='g', label='Aleatrio')  # Red line

# Labels and title
plt.xlabel('n')
#plt.xticks(np.arange(1, 100 + 1, 1))  # Set x-ticks from 1 to 20, stepping by 1
plt.ylabel(tipo)
plt.title(f"{nome} - {tipo}")

# Add legend and grid
plt.legend()  # Show the legend with the labels 'Fora Bruta' and 'Extra'
plt.grid(True)

plt.savefig(f"{nome}_metrics/{nome.lower()[0:-4]}_{tipo}.png", bbox_inches='tight')

# Show plot
plt.show()
```
