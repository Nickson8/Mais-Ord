```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import gamma
import pandas as pd

nome = "Heap"
tipo = "time"
df = pd.read_csv(f"./{nome}_metrics/{nome}_{tipo}_metrics.csv")

# Generate x values
x = np.linspace(1, 100000, 150)  # 1000 points between 1 and 20 for smoothness

# Calculate the factorial of each x value
y = [(i*np.log2(i))/250000000 for i in x]
y2 = [((i**2)/1.8) for i in x]
y_ord = [i for i in df["Ordered"]]
y_rev = [i for i in df["Reversed"]]
y_ale = [i for i in df["Randomized"]]

# Plotting
plt.plot(x, y, marker='', linestyle='-', color='b', label='Terica')  # Blue line
plt.plot(x, y_ord, marker='', linestyle='-', color='r', label='Ordenado')  # Red line
plt.plot(x, y_rev, marker='', linestyle='-', color='purple', label='Reverso')  # Red line
plt.plot(x, y_ale, marker='', linestyle='-', color='g', label='Aleatrio')  # Red line

# Labels and title
plt.xlabel('n')
#plt.xticks(np.arange(1, 100 + 1, 1))  # Set x-ticks from 1 to 20, stepping by 1
plt.ylabel(tipo)
plt.title(f"{nome} - {tipo}")

# Add legend and grid
plt.legend()  # Show the legend with the labels 'Fora Bruta' and 'Extra'
plt.grid(True)

plt.savefig(f"{nome}_metrics/{nome.lower()}_{tipo}.png", bbox_inches='tight')

# Show plot
plt.show()
```
